```python
from tkinter import*
from tkinter import messagebox
from tkinter import Tk, mainloop, TOP
import time
from tkinter.ttk import Combobox
import pandas as pd
import numpy as np
from PIL import ImageTk,Image
def left(event):       
    win1=Tk() 
    

    win1.title("Job Recommender")
    win1.iconbitmap('C:/Users/venka/Downloads/Aha-Soft-Standard-Portfolio-Unemployed.ico')
    Label(win1,text="OBJECTIVE").grid(row=0,column=0)
    objective="""Our Effort is to recommend jobs based on measuring similarity of other people profiles """
    Label(win1,text=objective).grid()
    win1.mainloop()
win=Tk() 
win.geometry('2000x2000') 
win.title("Job Recommender")
win.iconbitmap('C:/Users/venka/Downloads/Mat.ico')
my_img=ImageTk.PhotoImage(Image.open("C:/Users/venka/Downloads/job-logo-png-4.png"))
my_label=Label(image=my_img)
my_label.bind("<Button-1>",left)
my_label.grid(row=1000,column=1000)
label=Label(win,bg="LightBlue",fg="Black",text="Enter Highest Qualification",relief="raised",font="Helvetica 10 bold italic",
           )
label.grid(row=1,column=0)

hq=StringVar()
cc=StringVar()
ec=StringVar()
cp=StringVar()
ep=StringVar()
col=StringVar()
ind=StringVar()
scl=StringVar()
per=StringVar()
Tt=StringVar()
Ex=StringVar()
en=Entry(win,textvariable=hq)
en.grid(row=1,column=1)
label1=Label(win,bg="LightBlue",fg="Black",text="Enter Current Company",relief="raised",font="Helvetica 10 bold italic",
           )
label1.grid(row=2,column=0)
en1=Entry(win,textvariable=cc)
en1.grid(row=2,column=1)
label2=Label(win,bg="LightBlue",fg="Black",text="Enter Earlier Companies worked",relief="raised",font="Helvetica 10 bold italic",
           )
label2.grid(row=3,column=0)
en2=Entry(win,textvariable=ec)
en2.grid(row=3,column=1)
label3=Label(win,bg="LightBlue",fg="Black",text="Enter Current Position",relief="raised",font="Helvetica 10 bold italic",
           )
label3.grid(row=4,column=0)
en3=Entry(win,textvariable=cp)
en3.grid(row=4,column=1)
label4=Label(win,bg="LightBlue",fg="Black",text="Enter Earlier positions",relief="raised",font="Helvetica 10 bold italic",
           )
label4.grid(row=5,column=0)
en4=Entry(win,textvariable=ep)
en4.grid(row=5,column=1)
label5=Label(win,bg="LightBlue",fg="Black",text="Enter College",relief="raised",font="Helvetica 10 bold italic",
           )
label5.grid(row=6,column=0)
en5=Entry(win,textvariable=col)
en5.grid(row=6,column=1)
label6=Label(win,bg="LightBlue",fg="Black",text="Enter School",relief="raised",font="Helvetica 10 bold italic",
           )
label6.grid(row=7,column=0)
en6=Entry(win,textvariable=scl)
en6.grid(row=7,column=1)
label7=Label(win,bg="LightBlue",fg="Black",text="Enter Industry Knowledge",relief="raised",font="Helvetica 10 bold italic",
           )
label7.grid(row=8,column=0)
en7=Entry(win,textvariable=ind)
en7.grid(row=8,column=1)
label8=Label(win,bg="LightBlue",fg="Black",text="Enter Percentage",relief="raised",font="Helvetica 10 bold italic",
           )
label8.grid(row=9,column=0)
en8=Entry(win,textvariable=per)
en8.grid(row=9,column=1)
label9=Label(win,bg="LightBlue",fg="Black",text="Enter Tools and Technology you know",relief="raised",font="Helvetica 10 bold italic",
           )
label9.grid(row=10,column=0)
en9=Entry(win,textvariable=Tt)
en9.grid(row=10,column=1)
label10=Label(win,bg="LightBlue",fg="Black",text="Enter Experience",relief="raised",font="Helvetica 10 bold italic",
           )
en10=Entry(win,textvariable=Ex)
label10.grid(row=11,column=0)
en10.grid(row=11,column=1)
def recomm():
    win2=Tk()
    win2.title("Jobs Recommended")
    win2.iconbitmap('C:/Users/venka/Downloads/Aha-Soft-Standard-Portfolio-Unemployed.ico')
    df = pd.read_excel('profile_1.xlsx')
    df=df[['user_ID','Highest Qualification', 'Current Company',
    'Earlier Companies worked', 'Current Position', 'Earlier positions',
    'College', 'School', 'Industry Knowledge','Percentage','Tools and Technology','Experience']]
    features=['Highest Qualification', 'Current Company',
    'Earlier Companies worked', 'Current Position', 'Earlier positions',
    'College', 'School', 'Industry Knowledge','Percentage','Tools and Technology','Experience']
    for x in features:
        df[x]=df[x].fillna(' ')
    y=0
    a=en.get()
    b=en1.get()
    c=en2.get()
    d=en3.get()
    e=en4.get()
    f=en5.get()
    g=en6.get()
    h=en7.get()
    i=en8.get()
    j=en9.get()
    k=en10.get()
    new_row=({'user_ID':y,'Highest Qualification':a,'Current Company':b,'Earlier Companies worked':c,'Current Position':d,'Earlier positions':e,'College':f,'School':g,'Industry Knowledge':h,'Percentage':i,'Tools and Technology':j,'Experience':k})
    df=df.append(new_row,ignore_index=True,sort=False)
    def combine(row):
        return str(row['Highest Qualification'])+' '+str(row['Current Company'])+' '+str(row['Earlier Companies worked'])+' '+str(row['Current Position'])+' '+str(row['Earlier positions'])+' '+str(row['College'])+' '+str(row['School'])+' '+str(row['Industry Knowledge'])+' '+str(row['Percentage'])+' '+str(row['Tools and Technology'])+' '+str(row['Experience'])
    df["combined_features"]=df.apply(combine,axis=1)
    from sklearn.feature_extraction.text import TfidfVectorizer
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(df["combined_features"])
    from sklearn.metrics.pairwise import linear_kernel
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
    indices = pd.Series(df.index, index=df['user_ID']).drop_duplicates()
    def content_recommender(title, cosine_sim=cosine_sim, df=df, indices=indices):
        idx = indices[title]
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        j=1
        sim_scores = sim_scores[1:5]
        #course_indices = [i[0] for i in sim_scores]
        Label(win2,text="JOB",bg="LightBlue",relief="raised").grid(column=0,row=0)
        Label(win2,text="COMPANY",bg="LightBlue",relief="raised").grid(column=1,row=0)
        for i in sim_scores:
            Label(win2,text=df['Current Position'].iloc[i[0]]).grid(column=0,row=j)
            Label(win2,text=df['Current Company'].iloc[i[0]],underline=0).grid(column=1,row=j)
        #time.sleep(2)
            j=j+1
        
    content_recommender(0)
    win2.mainloop()
    


but=Button(win,text="Start",command=recomm,bg="Green",fg="White")
but.grid(row=15,column=0)
but2=Button(win,text="Exit",command=win.destroy,bg="Red",fg="White")
but2.grid(row=15,column=3)

     
win.mainloop()
```
