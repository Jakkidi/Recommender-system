```python
from tkinter import*
from tkinter import messagebox
from tkinter import Tk, mainloop, TOP
import time
from tkinter.ttk import Combobox
import pandas as pd
import numpy as np
from PIL import ImageTk,Image
def leftclick(event):       
    win=Tk() 
    win.geometry('600x600') 

    win.title("Online Course Recommender")
    win.iconbitmap('C:/Users/venka/Downloads/logo_umu_icon.ico')
    label=Label(win,bg="LightBlue",fg="Black",text="Enter The Course Name",relief="raised",font="Helvetica 10 bold italic",
           )
    label.grid(row=1,column=0)

    one=StringVar()
    en=Entry(win,textvariable=one)
    en.grid(row=1,column=1)
    def recomm():
        win2=Tk()
        win2.title("The Recommended Courses are")
        win2.iconbitmap('C:/Users/venka/Downloads/logo_umu_icon.ico')
        df = pd.read_csv('cour.csv')
        df=df[['Area','Title','Course site', 'Type', 'Level', 'description ',
       'skills gained','Link']]
        features=['Area','Title','Course site', 'Type', 'Level', 'description ',
       'skills gained','Link']
        for x in features:
            df[x]=df[x].fillna(' ')
        from sklearn.feature_extraction.text import TfidfVectorizer
        tfidf = TfidfVectorizer(stop_words='english')
        tfidf_matrix = tfidf.fit_transform(df)
        from sklearn.metrics.pairwise import linear_kernel
        cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
        indices = pd.Series(df.index, index=df['Title']).drop_duplicates()
        def content_recommender(title, cosine_sim=cosine_sim, df=df, indices=indices):
            idx = indices[title]
            sim_scores = list(enumerate(cosine_sim[idx]))
            sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
            j=0
            sim_scores = sim_scores[1:5]
            #course_indices = [i[0] for i in sim_scores]
            Label(win2,text="TITLE").grid(column=0,row=0)
            Label(win2,text="LINK").grid(column=1,row=0)
            Label(win2,text="LEVEL").grid(column=2,row=0)
            Label(win2,text="TYPE").grid(column=3,row=0)
            j=1
            for x in sim_scores:
                Label(win2,text=df['Title'].iloc[x[0]]).grid(column=0,row=j)
                Label(win2,text=df['Link'].iloc[x[0]]).grid(column=1,row=j)
                Label(win2,text=df['Level'].iloc[x[0]]).grid(column=2,row=j)
                Label(win2,text=df['Type'].iloc[x[0]]).grid(column=3,row=j)
                time.sleep(2)
                j=j+1
            

        content_recommender(en.get())
        win2.mainloop()
    but=Button(win,text="Start",command=recomm,bg="Green",fg="White")
    but.grid(row=10,column=0)
    but2=Button(win,text="Exit",command=win.destroy,bg="Red",fg="White")
    but2.grid(row=10,column=3)
    data=['Beginner','Intermediate','Advanced']
    cb=Combobox(win,values=data)
    cb.grid(row=5,column=0)
    Label(win,text="Select Level").grid(row=4,column=0)
    my_img=ImageTk.PhotoImage(Image.open("C:/Users/venka/Downloads/Rahul/New folder/logo.png"))
    my_label=Label(image=my_img)
    my_label.grid(row=0,column=3) 
    win.mainloop()
def read():
    root2=Tk()
    root2.title("Edu")
    root2.geometry('600x600')
    root2.iconbitmap('C:/Users/venka/Downloads/logo_umu_icon.ico')
    Label(root2,text="Edu Guiding Towards development",bg="LightBlue",fg="Black",relief="raised",font="Helvetica 10 bold italic").grid(row=0)
    Label(root2,text="Objectives:",fg="Black",bg="Green").grid(row=1)
    objective="""It guides new and existing learners to do courses based on their intrests.
It is a great platform for absolute beginners.
For absolute beginners it gives parallel recommendations and links of other websites.
Even for existing learners also it gives a parallel recommendations based on their working domain.
"""
    Label(root2,text=objective).grid(row=2)
    root2.mainloop()
root=Tk()
root.title("Edu")
one=StringVar()
root.iconbitmap('C:/Users/venka/Downloads/logo_umu_icon.ico')
myimg=ImageTk.PhotoImage(Image.open("C:/Users/venka/Downloads/Freedom-Education-Quotes-Wallpaper-10617.jpg"))
mylab=Label(image=myimg)
mylab.bind("<Button-1>",leftclick)
mylab.grid()
Button(root,text="Read More",command=read).grid()

root.mainloop()

```
